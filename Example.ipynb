```python
imput = df.drop('traffic',axis=1)
```

```python
# Apply MissForest imputation
imputer = MissForest()
```

```python

df_imputed = imputer.fit_transform(imput)
```

```python
df_imputed = pd.DataFrame(df_imputed, columns=imput.columns)
```

```python
df_imputed['traffic'] = df['traffic']
```

```python
df_imputed['origin_host'] = df_imputed['origin_host'].astype('int')
df_imputed['origin_port'] = df_imputed['origin_port'].astype('int')
df_imputed['response_host'] = df_imputed['response_host'].astype('int')
df_imputed['response_port'] = df_imputed['response_port'].astype('int')
```

```python
df_imputed.to_csv('/content/drive/MyDrive/Dataquest/train missfores imputed gpu.csv', index=False)
```
